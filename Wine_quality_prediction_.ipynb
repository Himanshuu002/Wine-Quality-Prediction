```python
# Define columns to check for outliers
columns_to_check = ['fixed acidity', 'volatile acidity', 'citric acid','residual sugar','chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol' ] # Updated column names

# Calculate Z-scores for each column
z_scores = np.abs((df[columns_to_check] - df[columns_to_check].mean()) / df[columns_to_check].std())

# Filter rows where any Z-score is greater than 3
outliers = df[(z_scores > 3).any(axis=1)]

# Remove outliers
df = df.drop(outliers.index)
```

```python
df.shape
```

## Data preprocessing

```python
# Separate data and lable
X=df.drop('quality',axis=1)
```

```python
X
```

## Lable binarization

```python

Y=df['quality'].apply(lambda y_value: 1 if y_value>=7 else 0)

# when quality is not good = 0
# when quality is good = 1
```

```python
print(Y)
```

## Train & Test Split

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.1,random_state=3)
```

```python
print(X.shape,x_train.shape,x_test.shape)
```

# Model Training:

```python
model_comp={}
```

```python
# Scale the training and test data

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

## 1. Logistic Regression





```python
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression(penalty='l2', max_iter=10000, random_state=42)
model1.fit(x_train_scaled,y_train)


from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

# Make predictions
y1 = model1.predict(x_test_scaled)
print(accuracy_score(y_test,y1))
print(f1_score(y_test,y1,average='weighted'))


model_comp['logistic regression']=[accuracy_score(y_test,y1),f1_score(y_test,y1,average='weighted')]
```

# 2. Random Forest Classifier

```python

from sklearn.ensemble import RandomForestClassifier
model2=RandomForestClassifier(n_estimators=300, random_state=42)
model2.fit(x_train,y_train)      ## remark: Scaled data is not necessary for random forest


from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

y2=model2.predict(x_test)
print(accuracy_score(y_test,y2))
print(f1_score(y_test,y2,average='weighted'))


model_comp['Random forest']=[accuracy_score(y_test,y2),f1_score(y_test,y2,average='weighted')]
```

## 3. Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
model3=DecisionTreeClassifier(max_depth=10, criterion="log_loss",splitter="best", min_samples_split=5,min_samples_leaf=3,max_features='sqrt')
model3.fit(x_train,y_train)


from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

y3=model3.predict(x_test)
print(accuracy_score(y_test,y3))
print(f1_score(y_test,y3,average='weighted'))


model_comp['decision tree']=[accuracy_score(y_test,y3),f1_score(y_test,y3,average='weighted')]
```

## 4. KNN

```python
from sklearn.neighbors import KNeighborsClassifier
model4=KNeighborsClassifier()
model4.fit(x_train_scaled,y_train)


from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

y4=model4.predict(x_test_scaled)
print(accuracy_score(y_test,y4))
print(f1_score(y_test,y4,average='weighted'))


model_comp['KNN']=[accuracy_score(y_test,y4),f1_score(y_test,y4,average='weighted')]
```

## 5. Support Vector Machine (SVM)

```python
from sklearn.svm import SVC
model5=SVC()
model5.fit(x_train_scaled,y_train)

from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

y5=model5.predict(x_test_scaled)
print(accuracy_score(y_test,y5))
print(f1_score(y_test,y5,average='weighted'))


model_comp['svc']=[accuracy_score(y_test,y5),f1_score(y_test,y5,average='weighted')]
```

## Model Comparsion

```python
df1=pd.DataFrame.from_dict(model_comp).T
df1.columns=['Accuracy','F1_score']
df1=df1.sort_values('F1_score',ascending=True)
df1.style.background_gradient(cmap='Reds')

```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import roc_curve, auc



models = ['logistic regression', 'Random forest', 'decision tree', 'KNN', 'svc']
accuracy_scores = [model_comp[model][0] for model in models]
f1_scores = [model_comp[model][1] for model in models]

# Create the plot
plt.figure(figsize=(10, 6))
X_axis = np.arange(len(models))

# Bar plot for Accuracy with percentage labels
bar1 = plt.bar(X_axis - 0.2, accuracy_scores, 0.4, label='Accuracy')
for rect in bar1:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width() / 2, height, f'{height:.2%}', ha='center', va='bottom')

# Bar plot for F1 Score with percentage labels
bar2 = plt.bar(X_axis + 0.2, f1_scores, 0.4, label='F1 Score')
for rect in bar2:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width() / 2, height, f'{height:.2%}', ha='center', va='bottom')

# Customize the plot
plt.xticks(X_axis, models, rotation=45, ha="right")
plt.ylabel("Scores")
plt.title("Comparison of Models")
plt.legend()
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

## Confusion Matrix

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay



# Logistic Regression Confusion Matrix
cm1 = confusion_matrix(y_test, y1)
disp1 = ConfusionMatrixDisplay(confusion_matrix=cm1)
disp1.plot(cmap='Greens')  # Changed color map to 'Greens'
plt.title('Logistic Regression Confusion Matrix')
plt.show()

# Random Forest Confusion Matrix
cm2 = confusion_matrix(y_test, y2)
disp2 = ConfusionMatrixDisplay(confusion_matrix=cm2)
disp2.plot(cmap='Reds')  # Changed color map to 'Reds'
plt.title('Random Forest Confusion Matrix')
plt.show()

# Decision Tree Confusion Matrix
cm3 = confusion_matrix(y_test, y3)
disp3 = ConfusionMatrixDisplay(confusion_matrix=cm3)
disp3.plot(cmap='Blues')  # Changed color map to 'Blues'
plt.title('Decision Tree Confusion Matrix')
plt.show()

# KNN Confusion Matrix
cm4 = confusion_matrix(y_test, y4)
disp4 = ConfusionMatrixDisplay(confusion_matrix=cm4)
disp4.plot(cmap='Oranges')  # Changed color map to 'Oranges'
plt.title('KNN Confusion Matrix')
plt.show()

# SVM Confusion Matrix
cm5 = confusion_matrix(y_test, y5)
disp5 = ConfusionMatrixDisplay(confusion_matrix=cm5)
disp5.plot(cmap='Purples')  # Changed color map to 'Purples'
plt.title('SVM Confusion Matrix')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc


models_pred_prob = {
    'Logistic Regression': model1.predict_proba(x_test_scaled)[:, 1],
    'Random Forest': model2.predict_proba(x_test)[:, 1],
    'Decision Tree': model3.predict_proba(x_test)[:, 1],
    'KNN': model4.predict_proba(x_test_scaled)[:, 1],
    'SVC': model5.decision_function(x_test_scaled) # decision_function for SVC
}

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))
fig.suptitle("ROC Curves for All Models", fontsize=16)


for i, (model_name, y_pred_prob) in enumerate(models_pred_prob.items()):
    fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
    roc_auc = auc(fpr, tpr)

    row = i // 3
    col = i % 3
    ax = axes[row, col]

    ax.plot(fpr, tpr, lw=2, label=f'{model_name} (AUC = {roc_auc:.2f})')
    ax.plot([0, 1], [0, 1], linestyle='--', lw=2, color='black', label='Random Guessing')
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_title(f'ROC Curve for {model_name}')
    ax.legend(loc="lower right")

if len(models_pred_prob) < 6:
    fig.delaxes(axes[1,2])

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
