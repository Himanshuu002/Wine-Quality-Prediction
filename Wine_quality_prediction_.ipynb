```python
# Define columns to check for outliers
columns_to_check = ['fixed acidity', 'volatile acidity', 'citric acid','residual sugar','chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol' ] # Updated column names

# Calculate Z-scores for each column
z_scores = np.abs((df[columns_to_check] - df[columns_to_check].mean()) / df[columns_to_check].std())

# Filter rows where any Z-score is greater than 3
outliers = df[(z_scores > 3).any(axis=1)]

# Remove outliers
data_no_outliers = df.drop(outliers.index)
z_scores = np.abs((df[columns_to_check] - df[columns_to_check].mean()) / df[columns_to_check].std())

# Filter rows where any Z-score is greater than 3
outliers = df[(z_scores > 3).any(axis=1)]

# Remove outliers
df = df.drop(outliers.index)
```

## Data preprocessing

```python
# Separate data and lable
X=df.drop('quality',axis=1)
```

```python
X
```

## Lable binarization

```python

Y=df['quality'].apply(lambda y_value: 1 if y_value>=7 else 0)

# when quality is not good = 0
# when quality is good = 1
```

```python
print(Y)
```

## Train & Test Split

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.1,random_state=3)
```

```python
print(X.shape,x_train.shape,x_test.shape)
```

# Model Training:

```python
model_comp={}
```

```python
# Scale the training and test data

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

## 1. Logistic Regression





```python
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression(penalty='l2', max_iter=10000, random_state=42)
model1.fit(x_train_scaled,y_train)


from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

# Make predictions
y1 = model1.predict(x_test_scaled)
print(accuracy_score(y_test,y1))
print(f1_score(y_test,y1,average='weighted'))


model_comp['logistic regression']=[accuracy_score(y_test,y1),f1_score(y_test,y1,average='weighted')]
```

# 2. Random Forest Classifier

```python

from sklearn.ensemble import RandomForestClassifier
model2=RandomForestClassifier(n_estimators=300, random_state=42)
model2.fit(x_train,y_train)      ## remark: Scaled data is not necessary for random forest


from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

y2=model2.predict(x_test)
print(accuracy_score(y_test,y2))
print(f1_score(y_test,y2,average='weighted'))


model_comp['Random forest']=[accuracy_score(y_test,y2),f1_score(y_test,y2,average='weighted')]
```

## 3. Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
model3=DecisionTreeClassifier(max_depth=10, criterion="log_loss",splitter="best", min_samples_split=5,min_samples_leaf=3,max_features='sqrt')
model3.fit(x_train,y_train)


from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

y3=model3.predict(x_test)
print(accuracy_score(y_test,y3))
print(f1_score(y_test,y3,average='weighted'))


model_comp['decision tree']=[accuracy_score(y_test,y3),f1_score(y_test,y3,average='weighted')]
```

## 4. KNN

```python
from sklearn.neighbors import KNeighborsClassifier
model4=KNeighborsClassifier()
model4.fit(x_train_scaled,y_train)


from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

y4=model4.predict(x_test_scaled)
print(accuracy_score(y_test,y4))
print(f1_score(y_test,y4,average='weighted'))


model_comp['KNN']=[accuracy_score(y_test,y4),f1_score(y_test,y4,average='weighted')]
```

## 5. Support Vector Machine (SVM)

```python
from sklearn.svm import SVC
model5=SVC()
model5.fit(x_train_scaled,y_train)

from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

y5=model5.predict(x_test_scaled)
print(accuracy_score(y_test,y5))
print(f1_score(y_test,y5,average='weighted'))


model_comp['svc']=[accuracy_score(y_test,y5),f1_score(y_test,y5,average='weighted')]
```

## Model Comparsion

```python
df1=pd.DataFrame.from_dict(model_comp).T
df1.columns=['Accuracy','F1_score']
df1=df1.sort_values('F1_score',ascending=True)
df1.style.background_gradient(cmap='Reds')
#print(df1)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier

# Example for Decision Tree:
param_grid = {
    'max_depth': [5, 10, 15, 20],
    'criterion': ['gini', 'entropy'],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(DecisionTreeClassifier(), param_grid, cv=5, scoring='accuracy')
grid_search.fit(x_train, y_train)

best_model = grid_search.best_estimator_
```

```python
# prompt: try ensemble learning and plot the accuracy

import matplotlib.pyplot as plt
from sklearn.ensemble import VotingClassifier

# Define the individual models
model1 = LogisticRegression(penalty='l2', max_iter=10000, random_state=42)
model2 = RandomForestClassifier(n_estimators=300, random_state=42)
model3 = DecisionTreeClassifier(max_depth=10, criterion="log_loss", splitter="best", min_samples_split=5, min_samples_leaf=3, max_features='sqrt')
model4 = KNeighborsClassifier()
model5 = SVC()

# Create the ensemble model using VotingClassifier
ensemble_model = VotingClassifier(estimators=[('lr', model1), ('rf', model2), ('dt', model3), ('knn', model4), ('svc', model5)], voting='hard') # voting='hard' for majority class voting

# Train the ensemble model
ensemble_model.fit(x_train_scaled, y_train)

# Make predictions
y_pred_ensemble = ensemble_model.predict(x_test_scaled)

# Evaluate the ensemble model
accuracy_ensemble = accuracy_score(y_test, y_pred_ensemble)
print(f"Ensemble Model Accuracy: {accuracy_ensemble}")

# Plot the accuracy of each model
model_names = ['Logistic Regression', 'Random Forest', 'Decision Tree', 'KNN', 'SVM', 'Ensemble']
accuracies = [accuracy_score(y_test, y1), accuracy_score(y_test, y2), accuracy_score(y_test, y3), accuracy_score(y_test, y4), accuracy_score(y_test, y5), accuracy_ensemble]


plt.figure(figsize=(10, 6))
plt.bar(model_names, accuracies, color=['skyblue', 'salmon', 'lightgreen', 'gold', 'plum', 'teal'])
plt.xlabel("Models")
plt.ylabel("Accuracy")
plt.title("Model Accuracy Comparison")
plt.ylim(0, 1) # Set y-axis limit to 0-1 for accuracy
plt.show()
```
